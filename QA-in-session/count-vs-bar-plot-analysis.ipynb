# Count vs Bar Plot Analysis

This notebook demonstrates the differences between count plots and bar plots using sample data and real-world sales data.

## Count Plot

A count plot is a type of bar plot where the height of each bar represents the count (frequency) of observations in each category.

### Use Cases:
- Ideal for categorical data where you want to show the frequency of different categories.
- Useful in exploratory data analysis to understand the distribution of categorical variables.

### Characteristics:
- X-axis represents the categories of the variable.
- Y-axis represents the count (frequency) of observations.
- Simple and easy to interpret.
- Limited to showing frequencies and does not display other metrics (like mean, sum, etc.).

## Bar Plot

A bar plot is a plot where the height of each bar can represent various summary statistics (e.g., mean, sum, count) of the variable.

### Use Cases:
- Useful for both categorical and numerical data.
- Can show various summary statistics, such as mean, median, sum, or count.
- Often used to compare different groups or categories.

### Characteristics:
- X-axis represents the categories or groups.
- Y-axis can represent various metrics (mean, sum, etc.).
- More flexible in showing different types of summaries.
- Can include error bars to show variability (e.g., standard deviation or confidence intervals).

## Real World Example: Sales Data

Let's use a real-world example to further illustrate the differences between count plots and bar plots.

### Sales Data

| Product    | Sales |
|------------|-------|
| Laptop     | 1000  |
| Tablet     | 700   |
| Smartphone | 1500  |
| Laptop     | 1100  |
| Tablet     | 750   |
| Smartphone | 1550  |
| Laptop     | 1150  |
| Tablet     | 800   |
| Smartphone | 1600  |

In this example, we'll create a count plot to show the number of each product sold and a bar plot to show the total sales for each product.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Real world example: Sales data
sales_data = {
    'Product': ['Laptop', 'Tablet', 'Smartphone', 'Laptop', 'Tablet', 'Smartphone', 'Laptop', 'Tablet', 'Smartphone'],
    'Sales': [1000, 700, 1500, 1100, 750, 1550, 1150, 800, 1600]
}
sales_df = pd.DataFrame(sales_data)
```

```python
sales_df
```

```python
# Create Count Plot for number of sold products
plt.figure(figsize=(10, 5))
sns.countplot(x='Product', data=sales_df)
plt.title('Count Plot of Products Sold')
plt.xlabel('Product')
plt.ylabel('Count')
plt.show();
```

<b>Observation:</b> We can see that this market has sold the same number of each product. <br>

```python
# Create Bar Plot for each product sales
plt.figure(figsize=(10, 5))
sns.barplot(x='Product', y='Sales', data=sales_df, estimator=sum)
plt.title('Bar Plot of Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show();
```

<b>Observation:</b> Here, we can see that the total sales of each product is different from the others, and it's clear that the total Sales of the 'Smartphne' is better than the other products, and more profitable for this market.

## Conclusion

- <b>Count plot</b> is used for <b>univariate</b> analysis.
- <b>Bar plot</b> is used for <b>bivariate</b> analysis.
- Every Count plot is Bar plot, but <b>not</b> every Bar plot is Count plot. (Count plot can be replaced with a Bar plot)
