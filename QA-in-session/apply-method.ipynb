```python
import pandas as pd

# Sample DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [6, 7, 8, 9, 10]}  # dictionary
df = pd.DataFrame(data)



df.head()
```

```python
# Define a custom function
def square(x):
    return x ** 2

# Apply the function to each element of a DataFrame column using apply()
df['A_squared'] = df['A'].apply(square)
```

```python
df['B_squared'] = df['B'].apply(square)
```

```python
df
```

```python
# Define a custom function that takes two arguments
def add_columns(x, y):
    return x + y

# Apply the function to two columns using apply() and a lambda function
df['Sum'] = df.apply(lambda row: add_columns(row['A'], row['A_squared']), axis=1)
```

```python
df
```

```python
# Define a custom function that takes a row as input
def add_columns(df):
    return df['A'] + df['B'] +df['A_squared']


```

```python
# Apply the function to two columns using apply() without lambda
columns=['A','B', 'A_squared'] #####
df['Sum_3'] = df[columns].apply(add_columns, axis=1)
```
